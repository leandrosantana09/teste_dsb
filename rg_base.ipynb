# 2.1 Analiosando o manual

2.1.1) item.csv

A tabela original contem 32 colunas, para uma primeira analise a interresante uma filtragem, para tal tarefa,\
será usado os seguintes criterios,

'CHAVES' descritas pelo manual (duas combinações diferentes):\
'NR_LICITACAO'\
'ANO_LICITACAO'\
'CD_TIPO_MODALIDADE'\
'NR_LOTE'\

'NR_ITEM_ORIGINAL'

Indicações do propio enunciado:\

'DS_ITEM'\
'VL_UNITARIO_HOMOLOGADO'


```python
base_item = base_item[['NR_LICITACAO','ANO_LICITACAO', 'CD_TIPO_MODALIDADE', 'NR_LOTE', 'NR_ITEM', 'NR_ITEM_ORIGINAL', 'DS_ITEM', 'VL_UNITARIO_HOMOLOGADO']]
base_item.head()
```

2.1.2) licitacao.csv

De acordo com o manual, existem duas chaves para a tabela 'Item', 

CHAVE 1:\
'NR_LICITACAO','ANO_LICITACAO', 'CD_TIPO_MODALIDADE', 'NR_LOTE', 'NR_ITEM_ORIGINAL'

CHAVE 2:\
'NR_LICITACAO','ANO_LICITACAO', 'CD_TIPO_MODALIDADE', 'NR_LOTE', 'NR_ITEM'

Assim sendo, será usada duas vezes o metodo 'drop_duplicates', resultado em dois dataframes, um para cada combinação de chave.

```python
base_chave_1 = base_item.drop_duplicates(subset=['NR_LICITACAO','ANO_LICITACAO', 'CD_TIPO_MODALIDADE', 'NR_LOTE', 'NR_ITEM_ORIGINAL'], 
                            #inplace=True, 
                            ignore_index=True)

base_chave_2 = base_item.drop_duplicates(subset=['NR_LICITACAO','ANO_LICITACAO', 'CD_TIPO_MODALIDADE', 'NR_LOTE', 'NR_ITEM'], 
                            #inplace=True, 
                            ignore_index=True)
                                                                                    
```

Usando o metodo concat, 'concatenamos' os dois dataframes, e em seguida será removido as linhas repididas,\
 fruto do produto dos dataframes.

```python
base_drop = pd.concat([base_chave_1, base_chave_2], 
axis=0,
)
base_item_final = base_drop.drop_duplicates(subset=['NR_LICITACAO','ANO_LICITACAO', 'CD_TIPO_MODALIDADE', 'NR_LOTE', 'NR_ITEM', 'NR_ITEM_ORIGINAL'], 
                            #inplace=True, 
                            ignore_index=True)
```

```python
base_licitacao = base_licitacao[['NR_LICITACAO','ANO_LICITACAO', 'CD_TIPO_MODALIDADE', 'NM_ORGAO_GERENCIADOR','DS_OBJETO', 'TP_OBJETO']] 
base_licitacao =  base_licitacao.loc[base_licitacao['TP_OBJETO']=='COM']
```

```python
base_licitacao.drop_duplicates(subset=['NR_LICITACAO', 'ANO_LICITACAO', 'CD_TIPO_MODALIDADE'], inplace=True)
```

# 3) Analise da 'base_unida'

```python
base_unida = base_licitacao.merge(base_item_final, on=['NR_LICITACAO', 'ANO_LICITACAO', 'CD_TIPO_MODALIDADE'], how='inner')
base_unida.head(5)
```
